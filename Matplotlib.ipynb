```python
# lineplot
# histogram
# Bar chart
# pie chart
# subplot
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
days = np.array([1,2,3,4,5])
temp = np.array([32,45,23,23,31])
plt.plot(days,temp)
```

#### by the analysing this graph we can see that day 2 is most hot day , and day 1 ,day5 have medium temperature and day3,4 have lowest temperature most changable point in this graph is in between day 2 to 3... we can see that temperatre was 45 in day 2 and in the next day temperature will decresed approx 25 means 45 to 25....

```python
days = np.array([1,2,3,4,5])
temp = np.array([32,45,23,23,31])
plt.xlabel('days')
plt.ylabel('temp')
plt.title('days vs temp graph')
plt.plot(days,temp)
```

### in this graph we only describe that in the x axis = days , and in y axis = temp and give a title......

```python
days = np.array([1,2,3,4,5])
temp1 = np.array([32,45,23,23,31])
temp2 = [33,42,25,26,30]
plt.xlabel('Days')
plt.ylabel("Temp.")
plt.title('Days vs Temp Graph')
plt.plot(days,temp1,color = 'r')
plt.plot(days,temp2)
```

### we describe 2 temperature in a single graph....

```python
days = np.array(['monday','tuesday','wednesday','thursday','friday','saturday','sunday'])
temp1 = np.array([22,23,24,25,26,28,30])
temp2 = np.array([22,24,26,20,30,32,28])
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Days vs Temp Graph')
plt.plot(days,temp1, color = 'g')
plt.plot(days,temp2, color = 'r')
```
